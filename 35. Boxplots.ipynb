```r
library(ggplot2)
df <- mtcars
```

```r
qplot(factor(cyl), mpg, data = mtcars, geom = "boxplot")
```

```r
pl <- ggplot(df,aes(x=cyl,y=mpg))
```

```r
pl + geom_boxplot()
```

```r
pl <- ggplot(df,aes(x=factor(cyl),y=mpg))
pl +geom_boxplot()
```

```r
pl <- ggplot(df,aes(x=factor(cyl),y=mpg))
pl +geom_boxplot() + coord_flip()
```

```r
pl <- ggplot(df,aes(x=factor(cyl),y=mpg))
pl +geom_boxplot(fill = 'blue')
```

```r
pl <- ggplot(df,aes(x=factor(cyl),y=mpg))
pl +geom_boxplot(aes(fill =factor(cyl))) 
```
