```r
df <- fread('Economist_Assignment_Data.csv',drop = 1)
```

```r
head(df)
```

```r
ggplot(df,aes(CPI,HDI)) + geom_point(aes(color= Region), alpha =0.45, size = 7,shape = 'o')
```

```r
ggplot(df,aes(CPI,HDI)) + geom_point(aes(color= Region), alpha =0.7, size = 4,shape =1)
```

```r
ggplot(df,aes(CPI,HDI)) + geom_point(aes(color= Region), alpha =0.7, size = 4,shape =1) + geom_smooth(aes(group=1))
```

```r
pl2 <- ggplot(df,aes(CPI,HDI)) + geom_point(aes(color= Region), alpha =0.7, size = 4,shape =1) + 
geom_smooth(aes(group=1),method = 'lm', formula = y~log(x), se = F , color = 'red')
pl2
```

```r
pl2 + geom_text(aes(label=Country, color = Region))
```

```r
pointsToLabel <- c("Russia", "Venezuela", "Iraq", "Myanmar", "Sudan",
                   "Afghanistan", "Congo", "Greece", "Argentina", "Brazil",
                   "India", "Italy", "China", "South Africa", "Spane",
                   "Botswana", "Cape Verde", "Bhutan", "Rwanda", "France",
                   "United States", "Germany", "Britain", "Barbados", "Norway", "Japan",
                   "New Zealand", "Singapore")

pl3 <- pl2 + geom_text(aes(label = Country), color = "gray20", 
                data = subset(df, Country %in% pointsToLabel),check_overlap = TRUE)

pl3
```

```r
library(ggthemes)
```

```r
pl4 <- pl3 + theme_economist_white()
pl4
```

```r
pl4 + scale_x_continuous(limits = c(.9,10.5),breaks = 1:10, name = 'CPI index')
```

```r
pl4 + ggtitle('Title!') + scale_x_continuous(limits = c(.9,10.5),breaks = 1:10, name = 'CPI index') 
+ scale_y_continuous(limits = c(0,1), name = 'HDI index')
```
