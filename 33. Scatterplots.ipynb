```r
library('ggplot2')
df <- mtcars
```

```r
head(df)
```

```r
qplot(wt,mpg,data=df)
```

```r
qplot(wt,mpg,data=df,color=cyl)
```

```r
qplot(wt,mpg,data=df,size=cyl)
```

```r
qplot(wt,mpg,data=df,size=cyl,color=cyl)
```

```r
qplot(wt,mpg,data=df,size=cyl,color=hp,alpha=0.6)
```

```r
p1 <- ggplot(df,aes(x=wt,y=mpg))
p1 + geom_point(size = 5 , alpha = 0.5)
```

```r
p1 + geom_point(aes(size = hp))
```

```r
p1 + geom_point(aes(size = cyl))
```

```r
# Because the cyl number is even, we can use factor to fix it!
p1 + geom_point(aes(size = factor(cyl)))
```

```r
p1 + geom_point(aes(shape = factor(cyl)),size =5)
```

```r
p1 + geom_point(aes(shape = factor(cyl),color = factor(cyl)),size =5)
```

```r
p1 + geom_point(aes(shape = factor(cyl),color =hp,size =5))
```

```r
p1 + geom_point(aes(shape = factor(cyl),color =hp,size =5)) + scale_color_gradient(low='blue',high='red')
```
